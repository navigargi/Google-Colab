```python
def split_seq(seq,  num_pieces):
  start = 0
  for i in range(num_pieces):
    stop  = start + len(seq) // num_pieces
    yield seq[start:stop]
    start = stop


def pearson(a1, a2):
  s1 = np.mean(a1)
  s2 = np.mean(a2)
  return np.sum((a1 - s1) * (a2 - s2)) / np.sqrt(np.sum((a1 - s1)**2) * np.sum((a2 - s2)**2))
```

```python
nvda_df = nvda_df.sort_values(by='Datetime')

lookback = 30
dates = nvda_df['Datetime']
prices = list(nvda_df['Open'].values)
counter_ = -1
price_series = []
for day in dates:
  counter_ += 1
  if counter_ >= lookback:
    price_series.append(prices[counter_ - lookback:counter_])

timeseries_df = pd.DataFrame(price_series)
```

```python
counter = 5
complexity = 5
for index, row in timeseries_df.iterrows():
  counter -= 1

  plt.plot([np.mean(r) for r in split_seq(list(row.values), complexity)])
  plt.grid()
  plt.show()
  if counter < 0:
    break
```

```python
complexity = 6
outcome_list = []
for index, row in timeseries_df.iterrows():
  simplified_values = []
  for r in split_seq(list(row.values), complexity):
    simplified_values.append(np.mean(r))
  correz = pearson(simplified_values, correlate_against)
  if correz > .5:
    outcome_list.append(1)
  else:
    outcome_list.append(0)
```

A simple moving average calculates the average price during a specified period of time.

A simple moving average is a technical indicator that can aid in determining if an asset price will continue or if it will reverse a bull or bear trend.

A simple moving average can be enhanced as an exponential moving average (EMA) that is more heavily weighted on recent price action

```python
data['EMA_100']=ta.EMA(data['High'], 5)
```

EMAs are designed to track price trends over specific time periods, such as 50 or 200 days.

Unlike simple moving averages, EMAs give greater weight to recent data.
Computing the EMA involves applying a multiplier to the simple moving average (SMA).

Moving average ribbons let traders see multiple EMAs at the same time.

```python
# prompt: can you you make a graph based on macd also using a histogram and join the two graphs and make it red for below 0 and make it green for above 0 and make the center line equal to 0 and make it from -10 to 10

import matplotlib.pyplot as plt
# Calculate MACD
macd, macd_signal, macd_hist = ta.MACD(data["High"], fastperiod=12, slowperiod=26, signalperiod=9)

# Create a figure and axes
fig, ax1 = plt.subplots(figsize=(10,6))

# Plot MACD and signal line
ax1.plot(macd, 'b-', label = "MACD")
ax1.plot(macd_signal, '--', color = "orange", label = "Signal")
ax1.set_ylim(-10, 10)  # Set y-axis limits
ax1.axhline(0, color='black', linewidth=0.5)  # Center line at 0

# Create a twin axis for the histogram
ax2 = ax1.twinx()

# Plot the histogram
colors = ['green' if val >= 0 else 'red' for val in macd_hist]
ax2.bar(macd_hist.index, macd_hist, color=colors, label = "Histogram")
ax2.set_ylim(-10, 10)  # Set y-axis limits to match MACD

# Set labels and title
ax1.set_xlabel("Date")
ax1.set_ylabel("MACD")
ax2.set_ylabel("Histogram")
plt.title("Nvidia")

# Display legend
fig.legend(loc="upper left")

plt.show()
```

Moving average convergence/divergence (MACD) is a technical indicator to help investors identify market entry points for buying or selling.

The MACD line is calculated by subtracting the 26-period exponential moving average (EMA) from the 12-period EMA.
The signal line is a nine-period EMA of the MACD line.

MACD is best used with daily periods, where the traditional settings of 26/12/9 days is the default.

```python
import yfinance as yf
from yahoo_fin import stock_info as si
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
from datetime import date


def get_live_price(ticker):
    data = yf.download(tickers = ticker, period= "1d") # Changed period to "1d" to get today's data
    df = data  # yf.download already returns a DataFrame
    today = date.today()
    value = df['Close'].iloc[-1] # Use iloc to get the last closing price
    return value


plt.style.use("dark_background")

stocks = ["AAPL", "NVDA", "TSLA"]
amounts = [20, 15, 20]

values = [get_live_price(stocks[i]) * amounts[i] for i in range(len(stocks))]
sectors = [yf.Ticker(x).get_info()['industry'] for x in stocks]
categories = [yf.Ticker(x).get_info()['category'] for x in stocks]

market_caps = [yf.Ticker(x).get_info()['marketCap'] for x in stocks]

cash = 100000

etfs = ["SPY", "TLT", "GLD", "BND", "VTI", "^OMX"]
etfs_amounts = [100, 20, 10, 40, 60, 100]
etf_values = [get_live_price(etfs[i])* etfs_amounts[i] for i in range(len(etfs))]

general_dist = {
    "Stocks": sum(values),
    "ETFS": sum(etf_values),
    "Cash": cash
}

sector_dist = {}
for i in range(len(sectors)):
    if sectors[i] not in sector_dist.keys():
        sector_dist[sectors[i]] = 0
    sector_dist[sectors[i]] += values[i]


category_dist = {}
for i in range(len(category)):
    if category[i] not in category_dist.keys():
        category_dist[categories[i]] = 0
    category_dist[categories[i]] += values[i]

market_caps_dist = {'small': 0.0, 'mid': 0.0, 'large': 0.0, 'huge': 0.0}
for i in range(len(stocks)):
    if market_caps[i] < 2000000000:
        market_caps_dist['small'] += values[i]
    elif market_caps[i] < 10000000000:
        market_caps_dist['mid'] += values[i]
    elif market_caps[i] < 1000000000000:
        market_caps_dist['large'] += values[i]
    else:
        market_caps_dist["huge"] += values[i]

# Corrected line to create subplots
fig, axs = plt.subplots(2, 2)

fig.suptitle("Portfolio Diversification Analysis", fontsize = 18) # Use suptitle for the figure title

# Replace 'color' with 'colors'
axs[0,0].pie(general_dist.values(), labels=general_dist.keys(), autopct="%1.1f%%", pctdistance = .8, colors = mcolors.TABLEAU_COLORS)
axs[0,0].set_title("General Distribution")
axs[0,1].pie(sector_dist.values(), labels=sector_dist.keys(), autopct="%1.1f%%", pctdistance = .8, colors = mcolors.TABLEAU_COLORS)
axs[0,1].set_title("Sector Distribution")
axs[1,0].pie(category_dist.values(), labels=category_dist.keys(), autopct="%1.1f%%", pctdistance = .8, colors = mcolors.TABLEAU_COLORS)
axs[1,0].set_title("Category Distribution")
axs[1,1].pie(market_caps_dist.values(), labels=market_caps_dist.keys(), autopct="%1.1f%%", pctdistance = .8, colors = mcolors.TABLEAU_COLORS)
axs[1,1].set_title("Market Caps Distribution")

```
