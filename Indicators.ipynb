```python
import datetime as dt
import talib as ta
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

```python
# You can replace NVDA with whatever ticker you want to analyze
data = yf.download(tickers="NVDA", period = "1y")
data
```

```python
# SMA
data['SMA_100'] = ta.SMA(data['High'], 10)
data
```

```python
# SMA Graph
plt.plot(data["High"], label="Price")
plt.plot(data["SMA_100"], label="SMA_100")
plt.legend()
plt.show()
```

```python
# EMA
data['EMA_100']=ta.EMA(data['High'], 5)
data
```

```python
# EMA Graph
plt.plot(data["High"], label = "Price")
plt.plot(data["EMA_100"], label = "EMA_100")
plt.legend()
plt.show()
```

```python
# RSI and the graph of RSI
data['RSI'] = ta.RSI(data['High'])

import matplotlib.pyplot as plt
# Separate RSI calculation and plotting
data['RSI'] = ta.RSI(data['High'])

# Create a figure and axis for RSI
fig, ax = plt.subplots(figsize=(12, 6))  # Increased figsize

# Plot RSI
ax.plot(data['RSI'], label='RSI')

# Plot overbought and oversold lines
ax.axhline(y=30, color='r', linestyle='-', label='Oversold (30)')
ax.axhline(y=70, color='g', linestyle='-', label='Overbought (70)')

# Set labels and title
ax.set_xlabel("Date")
ax.set_ylabel("RSI")
plt.title("RSI")

# Display legend
ax.legend()

plt.show()
```

```python
# MACD and the graph


import matplotlib.pyplot as plt
# Calculate MACD
macd, macd_signal, macd_hist = ta.MACD(data["High"], fastperiod=12, slowperiod=26, signalperiod=9)

# Create a figure and axes
fig, ax1 = plt.subplots(figsize=(10,6))

# Plot MACD and signal line
ax1.plot(macd, 'b-', label = "MACD")
ax1.plot(macd_signal, '--', color = "orange", label = "Signal")
ax1.set_ylim(-10, 10)  # Set y-axis limits
ax1.axhline(0, color='black', linewidth=0.5)  # Center line at 0

# Create a twin axis for the histogram
ax2 = ax1.twinx()

# Plot the histogram
colors = ['green' if val >= 0 else 'red' for val in macd_hist]
ax2.bar(macd_hist.index, macd_hist, color=colors, label = "Histogram")
ax2.set_ylim(-10, 10)  # Set y-axis limits to match MACD

# Set labels and title
ax1.set_xlabel("Date")
ax1.set_ylabel("MACD")
ax2.set_ylabel("Histogram")
plt.title("Nvidia")

# Display legend
fig.legend(loc="upper left")

plt.show()
```

```python
# Linear Regression and the graph

data["LINEARREG"] = ta.LINEARREG(data["High"], timeperiod=14)  # Example using talib

# Linear Regression plot
plt.figure(figsize=(10, 6))
plt.plot(data["High"], label="Price", color='blue')
plt.plot(data["LINEARREG"], label="Linear Regression", color='red')
plt.legend()
plt.title("Nvidia High Prices with Linear Regression")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
# All of the indicators combines besides the MACD

fig, axs = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]}, figsize = (10,6))
axs[0].plot(data["High"], color = 'b', label = "Price")
axs[0].plot(data["SMA_100"], color = 'g', label = "SMA")
axs[0].plot(data["EMA_100"], color = 'r', label = "EMA")
axs[0].plot(data["LINEARREG"], color = 'y', label = "Linear Reg")
axs[1].axhline(y=30, color='r', linestyle='-')
axs[1].axhline(y=70, color='g', linestyle='-')
axs[1].plot(data["RSI"], label = "RSI")

axs[0].legend()
axs[1].legend()
```
